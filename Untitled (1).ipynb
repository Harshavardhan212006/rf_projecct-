```python
import os
import cv2
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras import models
```

```python
dataset_path = r"C:\Users\Shobha P\OneDrive\Desktop\Dataset_LNA\Dataset_LNA"
```

```python
print(os.listdir(dataset_path))
```

```python
data = []

for lna_folder in sorted(os.listdir(dataset_path)):

    lna_path = os.path.join(dataset_path, lna_folder)

    if not os.path.isdir(lna_path):
        continue

    for variation in sorted(os.listdir(lna_path)):

        variation_path = os.path.join(lna_path, variation)

        if not os.path.isdir(variation_path):
            continue

        class_name = f"{lna_folder}_{variation}"

        for img in os.listdir(variation_path):

            if img.endswith((".png",".jpg",".jpeg")):

                data.append([
                    os.path.join(variation_path,img),
                    class_name
                ])

df = pd.DataFrame(data,columns=["filepath","label"])

print(df.head())
print("Total Images:",len(df))
```

```python
le = LabelEncoder()

df["label_encoded"] = le.fit_transform(df["label"])

NUM_CLASSES = len(le.classes_)

print(NUM_CLASSES)
```

```python
train_df, temp_df = train_test_split(
    df,
    test_size=0.30,
    stratify=df["label_encoded"],
    random_state=42
)

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.50,
    stratify=temp_df["label_encoded"],
    random_state=42
)

print(len(train_df))
print(len(val_df))
print(len(test_df))
```

```python
IMG_SIZE = 224
BATCH_SIZE = 128
AUTOTUNE = tf.data.AUTOTUNE
```

```python
def process_image(path,label):

    image = tf.io.read_file(path)

    image = tf.io.decode_png(
        image,
        channels=3
    )

    image = tf.image.resize(
        image,
        (IMG_SIZE,IMG_SIZE)
    )

    image = tf.cast(
        image,
        tf.float32
    ) / 255.0

    label = tf.one_hot(
        label,
        NUM_CLASSES
    )

    return image,label
```

```python
data_augmentation = tf.keras.Sequential([

    layers.RandomBrightness(0.1),

    layers.RandomContrast(0.1)

])
```

```python
def create_train_dataset(df):

    ds = tf.data.Dataset.from_tensor_slices(
        (
            df["filepath"].values,
            df["label_encoded"].values
        )
    )

    ds = ds.shuffle(len(df))

    ds = ds.map(
        process_image,
        num_parallel_calls=AUTOTUNE
    )

    ds = ds.map(
        lambda x,y:(data_augmentation(x),y),
        num_parallel_calls=AUTOTUNE
    )

    ds = ds.batch(BATCH_SIZE)

    ds = ds.prefetch(AUTOTUNE)

    return ds


def create_test_dataset(df):

    ds = tf.data.Dataset.from_tensor_slices(
        (
            df["filepath"].values,
            df["label_encoded"].values
        )
    )

    ds = ds.map(
        process_image,
        num_parallel_calls=AUTOTUNE
    )

    ds = ds.batch(BATCH_SIZE)

    ds = ds.prefetch(AUTOTUNE)

    return ds
```

```python
train_ds = create_train_dataset(train_df)
```

```python
AUTOTUNE = tf.data.AUTOTUNE

def create_train_dataset(df):

    ds = tf.data.Dataset.from_tensor_slices(
        (
            df["filepath"].values,
            df["label_encoded"].values
        )
    )

    ds = ds.shuffle(
        len(df),
        reshuffle_each_iteration=True
    )

    ds = ds.map(
        process_image,
        num_parallel_calls=AUTOTUNE
    )

    ds = ds.batch(128)

    ds = ds.prefetch(AUTOTUNE)

    return ds


def create_test_dataset(df):

    ds = tf.data.Dataset.from_tensor_slices(
        (
            df["filepath"].values,
            df["label_encoded"].values
        )
    )

    ds = ds.map(
        process_image,
        num_parallel_calls=AUTOTUNE
    )

    ds = ds.batch(128)

    ds = ds.prefetch(AUTOTUNE)

    return ds
```

```python
train_ds = create_train_dataset(train_df)

val_ds = create_test_dataset(val_df)

test_ds = create_test_dataset(test_df)
```

```python
model = models.Sequential([

    layers.Input(shape=(224,224,3)),

    layers.Conv2D(64,(3,3),padding='same',activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D(),

    layers.Conv2D(128,(3,3),padding='same',activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D(),

    layers.Conv2D(256,(3,3),padding='same',activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D(),

    layers.Conv2D(512,(3,3),padding='same',activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D(),

    layers.GlobalAveragePooling2D(),

    layers.Dense(512,activation='relu'),
    layers.Dropout(0.4),

    layers.Dense(256,activation='relu'),
    layers.Dropout(0.3),

    layers.Dense(
        NUM_CLASSES,
        activation='softmax'
    )
])

model.summary()
```

```python
callbacks = [

    tf.keras.callbacks.ModelCheckpoint(

        "best_rf_cnn.keras",

        save_best_only=True,

        monitor='val_accuracy'
    ),

    tf.keras.callbacks.EarlyStopping(

        monitor='val_accuracy',

        patience=10,

        restore_best_weights=True
    ),

    tf.keras.callbacks.ReduceLROnPlateau(

        monitor='val_loss',

        factor=0.5,

        patience=4
    )
]
```

```python
optimizer = tf.keras.optimizers.AdamW(
    learning_rate=1e-4,
    weight_decay=1e-5
)

model.compile(
    optimizer=optimizer,
    loss='categorical_crossentropy',
    metrics=[
        'accuracy',
        tf.keras.metrics.TopKCategoricalAccuracy(
            k=5,
            name='top5_acc'
        )
    ]
)
```

```python
history = model.fit(

    train_ds,

    validation_data=val_ds,

    epochs=50,

    callbacks=callbacks
)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.legend(['Train','Val'])

plt.title("Accuracy")

plt.subplot(1,2,2)

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.legend(['Train','Val'])

plt.title("Loss")

plt.show()
```

```python
results = model.evaluate(test_ds)

print(results)
```
